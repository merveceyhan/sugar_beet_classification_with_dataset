```python
from sklearn.preprocessing import StandardScaler
# https://www.veribilimiokulu.com/naive-bayes-yontemiyle-siniflandirma-classification-with-naive-bayes-python-ile-uygulama/

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

### Logistic Regression

```python
from sklearn.linear_model import LinearRegression

reg_model = LinearRegression()  
reg_model.fit(X_train, y_train)
```

```python
y_pred_reg= reg_model.predict(X_test)

print("Prediction for test set: {}".format(y_pred_reg))
```

```python
# calculate MAE, MSE, RMSE
import sklearn.metrics as metrics

print(metrics.mean_absolute_error(y_test, y_pred_reg))
print(metrics.mean_squared_error(y_test, y_pred_reg))
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred_reg)))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_reg.round()))
```

### KNN

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor()
```

```python
knn_model.fit(X_train, y_train)
```

```python
y_pred_knn= knn_model.predict(X_test)

print("Prediction for test set: {}".format(y_pred_knn))
```

```python
import sklearn.metrics as metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_knn))

print(metrics.mean_absolute_error(y_test, y_pred_knn))
print(metrics.mean_squared_error(y_test, y_pred_knn))
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred_knn)))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_knn.round()))
```
